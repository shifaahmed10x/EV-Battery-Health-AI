```python
import pandas as pd
import joblib

# Load dataset
data = pd.read_csv("../data/battery_data.csv")

# Same features used in training
features = [
    'Voltage_measured',
    'Current_measured',
    'Temperature_measured',
    'Current_charge',
    'Voltage_charge'
]

X = data[features]

# Load trained model
model = joblib.load("../models/soh_model.pkl")
```

```python
import shap

# Create explainer
explainer = shap.Explainer(model, X)

# Explain predictions
shap_values = explainer(X)
```

```python
shap.summary_plot(shap_values, X)
```

```python
shap.plots.waterfall(shap_values[0])
```
